```python
import pickle
polygon = pickle.load(open('./data/HangJeongDongPolygon.pickle', 'rb'))
from sympy import Point, Polygon
import pandas as pd
from tqdm import tqdm
df = pd.read_csv('./data/concat.csv')
object_id = []
for row in tqdm(df.loc):
    point = Point(row['HPPN_X'], row['HPPN_Y'])
    enclose_id = -1
    for id, ps in polygon.items():
        for p in ps:
            if p.encloses_point(point):
                enclose_id = id
                break
        if enclose_id != -1:
            break
    print(enclose_id)
    object_id.append(enclose_id)
df['OBJECT_ID'] = object_id
df.to_csv('NPA2020_object_id.csv', index=False)
```
