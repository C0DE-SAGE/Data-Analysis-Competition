```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
kp_acc = pd.read_csv('./data/kp_acc_addr.csv')
npa_acc = pd.read_csv('./data/npa_acc_addr.csv')
```

```python
kp_acc
```

```python
kp_acc['ACC_PLACE'] = kp_acc['HPPN_PNU_ADDR'].apply(lambda x : x.split(' ')[0])
kp_acc.loc[kp_acc['ACC_PLACE'] == "충남", 'ACC_PLACE'] = "충청남도"
kp_acc.loc[kp_acc['ACC_PLACE'] == "대전", 'ACC_PLACE'] = "대전광역시"
```

```python
npa_acc['ACC_PLACE'] = npa_acc['HPPN_OLD_ADDR'].apply(lambda x : x.split(' ')[0])
npa_acc.loc[npa_acc['ACC_PLACE'] == "충남", 'ACC_PLACE'] = "충청남도"
npa_acc.loc[npa_acc['ACC_PLACE'] == "대전", 'ACC_PLACE'] = "대전광역시"
```

```python
kp_acc=kp_acc[kp_acc['ACC_PLACE'].isin(['충청남도', '대전광역시', '세종특별자치시'])].reset_index().drop(['index'], axis=1)
npa_acc=npa_acc[npa_acc['ACC_PLACE'].isin(['충청남도', '대전광역시', '세종특별자치시'])].reset_index().drop(['index'], axis=1)
```

```python
kp_acc.drop(['RECV_DEPT_NM'], axis=1, inplace=True)
kp_acc.rename(columns={'HPPN_PNU_ADDR':'HPPN_OLD_ADDR'}, inplace=True)
```

```python
acc_concat = pd.concat([kp_acc,npa_acc],axis=0)
acc_concat
```

```python
acc_concat.to_csv('./data/concat.csv', index=False)
```
