```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
VOICE_CATEGORY = [215]
ACC_CATEGORY = [401,402,403,406]
PATH = './data'
manuel = PATH + '/codeBook_v3.csv'
kp2020_path = PATH + '/KP2020.csv'
kp2021_path = PATH + '/KP2021.csv'
npa2020_path = PATH + '/NPA2020.csv'
```

```python
npa2020 = pd.read_csv(npa2020_path, encoding='cp949')
kp2021 = pd.read_csv(kp2021_path, encoding='cp949')
kp2020 = pd.read_csv(kp2020_path, encoding='cp949')
```

```python
npa2020
```

```python
from datetime import datetime
def get_unix(df):
    year = df['RECV_CPLT_DT'][2:4]
    month = df['RECV_CPLT_DT'][4:6]
    day = df['RECV_CPLT_DT'][6:]
    time = df['RECV_CPLT_TM'].zfill(6)
    time_format = year + '/' + month + '/' + day + ' ' + time[:2] + ':' + time[2:4] + ':' + time[4:6]
    unix = datetime.strptime(time_format,"%y/%m/%d %H:%M:%S").timestamp()
    return unix
```

```python
npa2020['RECV_CPLT_DT'] = npa2020['RECV_CPLT_DT'].astype(str)
npa2020['RECV_CPLT_TM'] = npa2020['RECV_CPLT_TM'].astype(str)
```

```python
npa2020['RECV_CPLT_DM'] = npa2020.apply(get_unix,axis=1)
```

```python
'''Memory reduce'''
start_memory = npa2020.memory_usage().sum()

npa2020['HPPN_OLD_ADDR'].fillna(0, inplace=True)
npa2020.drop('SME_EVT_YN', axis=1, inplace=True)
npa2020.dropna(inplace=True)

npa2020['HPPN_X'] = npa2020['HPPN_X'].astype(np.float32)
npa2020['HPPN_Y'] = npa2020['HPPN_Y'].astype(np.float32)
npa2020['RECV_CPLT_DM'] = npa2020['RECV_CPLT_DM'].astype(np.uint32)
npa2020['NPA_CL'] = npa2020['NPA_CL'].astype(np.uint8)
npa2020['EVT_STAT_CD'] = npa2020['EVT_STAT_CD'].astype(np.uint8)
npa2020['EVT_CL_CD'] = npa2020['EVT_CL_CD'].astype(np.uint16) 
npa2020.loc[npa2020['RPTER_SEX']=='{', 'RPTER_SEX'] = 3
npa2020.loc[npa2020['RPTER_SEX']==' ', 'RPTER_SEX'] = 3
npa2020['RPTER_SEX'].fillna(3, inplace=True)
npa2020['RPTER_SEX'] = npa2020['RPTER_SEX'].astype(np.uint8)


end_memory = npa2020.memory_usage().sum()
print(f"Memory usage of dataframe after reduction {end_memory} MB")
print(f"Reduced by {100 * (start_memory - end_memory) / start_memory} % ")
```

```python
npa2020 = npa2020[[list(npa2020.columns)[-1]] + list(npa2020.columns)[2:-1]]
```
