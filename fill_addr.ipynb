```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
kp_acc = pd.read_csv('./data/kp_acc.csv')
npa_acc = pd.read_csv('./data/npa_acc.csv')
npa_acc=npa_acc.loc[npa_acc['HPPN_X'] != 0.0]
kp_acc_none_loca_addr = pd.read_csv('./data/kp_acc_none_loca_addr.csv')
npa_acc_none_loca_addr = pd.read_csv('./data/npa_acc_none_loca_addr.csv')
```

```python
kp_acc_none_loca = kp_acc.loc[kp_acc['HPPN_PNU_ADDR'] == '0'].reset_index()
kp_acc_none_loca
```

```python
npa_acc_none_loca = npa_acc.loc[npa_acc['HPPN_OLD_ADDR'] == '0'].reset_index()
npa_acc_none_loca
```
